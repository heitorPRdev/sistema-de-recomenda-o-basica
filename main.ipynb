```python
import pandas as pd
df = pd.read_csv("tabela_recomendacao_600_linhas.csv")
x = df[["Acao","Aventura","Documentario","Documentario","SciFi","Terror"]]
y = df[["Filme_Recomendado"]]
x
```

```python
from sklearn.model_selection import train_test_split
x_train, X_test, y_train, Y_teste = train_test_split(x,y,test_size=0.4,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier()
tree.fit(x_train,y_train)
pred = tree.predict(X_test)
pred
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = ((12, 5)))
plt.subplot(1,3,1)
plt.hist(X_test,orientation="vertical")

plt.legend(["Ação", "Aventura","Documentario","Scifi", "Terror"],loc='lower center',
           fancybox=True,shadow=True, borderpad=1,framealpha=1)

plt.subplot(1,2,2)
plt.hist(pred,orientation="horizontal",align="right")
plt.show()
```
