```python
import pandas as pd
df = pd.read_csv("tabela_recomendacao_600_linhas.csv")
x = df[["Acao","Aventura","Documentario","Documentario","SciFi","Terror"]]
y = df[["Filme_Recomendado"]]
```

```python
from sklearn.model_selection import train_test_split
x_train, X_test, y_train, Y_teste = train_test_split(x,y,test_size=0.4,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier()
tree.fit(x_train,y_train)
pred = tree.predict(X_test)
pred
```

```python
import matplotlib.pyplot as plt
plt.hist(Y_teste,orientation="horizontal")

plt.show()
```
